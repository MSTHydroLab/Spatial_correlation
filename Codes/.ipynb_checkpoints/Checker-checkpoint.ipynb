```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


def plot_grid_selection(
    grid_id,
    grid_df,
    stations_df,
    nearest7_df,
    stn_x_col="NAD83_15N_Long",
    stn_y_col="NAD83_15N_Lat",
    grid_x_col="UTM_Easting",
    grid_y_col="UTM_Northing",
    max_radius_km=10,
):
    """
    Visual diagnostic plot for one grid cell.
    """

    # ---- Extract grid center
    row = grid_df[grid_df["id"] == grid_id].iloc[0]
    tx = float(row[grid_x_col])
    ty = float(row[grid_y_col])

    # ---- Extract selected gauges
    sel_row = nearest7_df[nearest7_df["id"] == grid_id].iloc[0]
    selected_ids = [sel_row[f"g{i}"] for i in range(1, 8)]
    selected_ids = [
        int(x) for x in selected_ids
        if not pd.isna(x)
    ]

    # ---- All gauges within 10 km
    gx = stations_df[stn_x_col].values
    gy = stations_df[stn_y_col].values
    gid = stations_df["ID"].astype(int).values

    dx = gx - tx
    dy = gy - ty
    dist = np.sqrt(dx**2 + dy**2)

    mask = dist <= max_radius_km * 1000

    # ---- Plot
    fig, ax = plt.subplots(figsize=(7, 7))

    # All gauges within 10 km
    ax.scatter(gx[mask], gy[mask], s=20, alpha=0.4, label="All gauges ≤10km")

    # Selected gauges
    sel_mask = np.isin(gid, selected_ids)
    print("Matched:", sel_mask.sum(), "out of", len(selected_ids))
    ax.scatter(
        gx[sel_mask],
        gy[sel_mask],
        s=160,
        color="red",
        edgecolor="black",
        linewidth=1.5,
        zorder=5,
        label="Selected 7",
    )

    # Grid center
    ax.scatter(tx, ty, s=120, marker="^", color="black", label="Grid center")

    # Radial lines
    for i in np.where(sel_mask)[0]:
        ax.plot([tx, gx[i]], [ty, gy[i]], linewidth=1)

    # Circles
    circle5 = plt.Circle((tx, ty), 5000, fill=False, linestyle="--", color="gray")
    circle10 = plt.Circle((tx, ty), 10000, fill=False, linestyle=":", color="gray")

    ax.add_patch(circle5)
    ax.add_patch(circle10)

    ax.set_aspect("equal")
    ax.set_title(f"Grid ID {grid_id}")
    ax.set_xlabel("UTM Easting (m)")
    ax.set_ylabel("UTM Northing (m)")
    ax.legend()
    plt.tight_layout()
    plt.show()
```

```python
grid_df = pd.read_csv(
    "/mnt/12TB/Sujan/Spatial_correlation/Codes/01_Correlation_and_Variogram/grid_centers_full.csv"
)

stations_df = pd.read_csv(
    "/mnt/12TB/Sujan/Spatial_correlation/Codes/01_Correlation_and_Variogram/Stations_df.csv"
)

nearest7_df = pd.read_csv(
    "/mnt/12TB/Sujan/Spatial_correlation/Codes/dependent_files/grid_nearest7.csv"
)

plot_grid_selection(
    grid_id=12391609,
    grid_df=grid_df,
    stations_df=stations_df,
    nearest7_df=nearest7_df
)
```
